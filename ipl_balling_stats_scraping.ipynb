## 3.Data scraping

- Scrape bowling stats for players from below website :
Website url : https://www.iplt20.com/stats/2022/

```python
import requests
from urllib.request import urlopen as uReq
from selenium import webdriver
import time
import warnings
warnings.filterwarnings("ignore")
import pandas as pd

url = "https://www.iplt20.com/stats/2022/"
# initiating the webdriver.
driver = webdriver.Chrome('./chromedriver')
driver.get(url)

# load all data by clicking 'View All' button
btn_click = driver.find_elements_by_css_selector(".recordtab__btn")

for u in btn_click:
    if u.text == "Bowling":
        u.click()
        time.sleep(1)
        # load all data by clicking 'View All' button
        btn_click = driver.find_element_by_css_selector(".np-mostrunsTab__btn a")
        #perform click with execute_script
        driver.execute_script("arguments[0].click();", btn_click);
        
        table2 = driver.find_element_by_css_selector(".np-mostrunsTableBg table.np-mostruns_table")
        
        all_data = table2.text
        
        # collecting the data
        all_li, player_li = [], []
        row_len = len(all_data.split('\n'))

        for i in range(3,row_len,3):
            all_li.append(all_data.split('\n')[i].split(' '))

        for i in range(2,row_len,3):
            player_li.append(all_data.split('\n')[i])

        table_cols = all_data.split('\n')[0].split(' ')

        # inseting data in dataframe
        df_ipl = pd.DataFrame(all_li, columns = table_cols[2:])
        df_ipl['Player'] = player_li

        # shift column 'Player' to first position
        first_column = df_ipl.pop('Player')
        df_ipl.insert(0, 'Player', first_column)

        df_ipl.to_csv(r'ipl_bowling_stats.csv', index = False)
        driver.close()
df_ipl.head(10)
```
